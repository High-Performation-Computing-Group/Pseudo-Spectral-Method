```python
from IPython.display import display,Math,Latex
import numpy as np
from mpl_toolkits import mplot3d

import matplotlib.pyplot as plt
import scipy.fftpack as sci
```

```python
Nx=64
Ny=64
t_min=0.00
t_max=2.00
dt=.000100
mu=0.000
a=2*np.pi
x=np.linspace(0,a,Nx)
y=np.linspace(0,a,Ny)
Ux=[[0 for i in range(Ny)] for j in range(Nx)]
dux=[[0 for i in range(Ny)] for j in range(Nx)]
phik=[[complex(0,1) for i in range(Ny)] for j in range(Nx)]
# Uk=[[complex(0,1) for i in range(Ny)] for j in range(Nx)]
k=1
for i in range(0,Nx):
    for j in range(0,Ny):
        Ux[i][j]=(np.sin(x[i])*np.cos(y[j]))/float(2)
Ukx=sci.fft(Ux)
for i in range(0,Nx):
    if i<= Nx/2:
        kx=2.00*(np.pi)*float(i)/a
    else:
        kx=2.00*(np.pi)*float(i-Nx)/a
    for j in range(0,Ny):
        if j<= Nx/2:
            ky=2.00*(np.pi)*float(j)/a
        else:
            ky=2.00*(np.pi)*float(j-Ny)/a
        if(i==0 and j==0):
            phik[i][j]=0.00
        else:
            phik[i][j]=Ukx[i][j]/(kx*kx+ky*ky)
phi=sci.ifft(phik).real
            
        
```
